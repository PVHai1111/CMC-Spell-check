```python
import os
```

```python
input_file = r'C:\Users\Admin\Downloads\data.json'
output_file_prefix = 'part_'
chunk_size = 50 * 1024 * 1024  # 50MB in bytes
```

```python
def split_file(input_file, output_file_prefix, chunk_size):
    file_number = 0
    with open(input_file, 'r', encoding='utf-8') as infile:
        while True:
            chunk_data = infile.read(chunk_size)
            if not chunk_data:
                break
            output_file = f"{output_file_prefix}{file_number}.json"
            with open(output_file, 'w', encoding='utf-8') as outfile:
                outfile.write(chunk_data)
            file_number += 1
            print(f"Written chunk {file_number} to {output_file}")
```

```python
split_file(input_file, output_file_prefix, chunk_size)
```
